```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de entrenamiento (50 puntos en [0, 2π])
np.random.seed(42)  # Para reproducibilidad
x_train = np.linspace(0, 2 * np.pi, 50)
y_train = np.sin(x_train)

# Inicialización de parámetros del Adaline
w = np.random.randn() * 0.01  # Peso (inicial pequeño)
b = 0.0                       # Sesgo
learning_rate = 0.01           # Tasa de aprendizaje
epochs = 1000                  # Número de épocas

# Entrenamiento del Adaline
errors = []
for epoch in range(epochs):
    total_error = 0
    for x, y_true in zip(x_train, y_train):
        # Predicción (salida lineal)
        y_pred = w * x + b
        
        # Error (diferencia cuadrática)
        error = y_true - y_pred
        total_error += error ** 2
        
        # Actualizar peso y sesgo (regla delta)
        w += learning_rate * error * x
        b += learning_rate * error
    
    # Guardar el error medio por época
    errors.append(total_error / len(x_train))
    
    # Opcional: Imprimir progreso cada 100 épocas
    if epoch % 100 == 0:
        print(f"Época {epoch}, Error: {errors[-1]:.4f}")

# Predicción en datos de prueba (para visualización)
x_test = np.linspace(0, 2 * np.pi, 100)
y_test = np.sin(x_test)
y_pred = w * x_test + b

# Gráfica de resultados
plt.figure(figsize=(10, 6))
plt.plot(x_test, y_test, label="Función real: $\sin(x)$", color="blue")
plt.plot(x_test, y_pred, label="Predicción ADALINE", linestyle="--", color="red")
plt.scatter(x_train, y_train, label="Datos de entrenamiento", color="green", alpha=0.5)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Aproximación de $\sin(x)$ con ADALINE")
plt.legend()
plt.grid(True)
plt.show()

# Gráfica del error durante el entrenamiento
plt.figure(figsize=(10, 4))
plt.plot(range(epochs), errors, label="Error cuadrático medio")
plt.xlabel("Época")
plt.ylabel("Error")
plt.title("Error durante el entrenamiento")
plt.grid(True)
plt.show()

print(f"Parámetros finales: w = {w:.4f}, b = {b:.4f}")
```
